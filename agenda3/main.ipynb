```python
import os
from stable_baselines3 import PPO
from stable_baselines3.common.vec_env import DummyVecEnv, VecMonitor
from env import TradingEnv
from extractor import CNNPolicy
from pathlib import Path
import pandas as pd
import numpy as np

# --- Config ---
meta_path = Path("../reinforcement_learning/dataset/meta.csv").resolve()
meta_df = pd.read_csv(meta_path, parse_dates=["timestamp"])
meta_root = meta_path.parent.resolve()

# Split
split_ratio = 0.8
split_idx = int(len(meta_df) * split_ratio)
train_df = meta_df.iloc[:split_idx].reset_index(drop=True)
test_df = meta_df.iloc[split_idx:].reset_index(drop=True)

print(f"[INFO] Training on {len(train_df)} samples, testing on {len(test_df)}")

# --- Env builders ---
def make_train_env():
    def _init():
        return TradingEnv(
            meta_df=train_df,
            root_dir=meta_root,
            image_size=(128, 128),
            starting_balance=100_000,
            leverage=1,
            risk_per_trade=1,
            log_path="./results/trading_log_v1_1t1_train.csv"
        )
    return _init

def make_test_env():
    def _init():
        return TradingEnv(
            meta_df=test_df,
            root_dir=meta_root,
            image_size=(128, 128),
            starting_balance=100_000,
            leverage=1,
            risk_per_trade=1,
            log_path="./results/trading_log_v1_1t1_test.csv"
        )
    return _init

# --- Env setup ---
train_env = DummyVecEnv([make_train_env()])
train_env = VecMonitor(train_env)

test_env = DummyVecEnv([make_test_env()])

# --- Model path ---
base_model_path = "./results/ppo-cnn-trading"

# --- Load or initialize model ---
if os.path.exists(base_model_path + ".zip"):
    print(f"[üìÇ LOADING EXISTING MODEL from {base_model_path}]")
    model = PPO.load(base_model_path, env=train_env, device="cuda")
else:
    print("[üÜï INITIALIZING NEW MODEL]")
    model = PPO(
        policy=CNNPolicy,
        env=train_env,
        verbose=1,
        n_steps=256,
        batch_size=64,
        learning_rate=1e-4,
        tensorboard_log="./results/tensorboard_logs",
        device="cuda"
    )

# --- Training loop ---
total_target_steps = 1_000_000
chunk_size = 100_000
chunks = total_target_steps // chunk_size

for i in range(chunks):
    print(f"\n[üîÅ TRAINING CHUNK {i+1}/{chunks}]")
    model.learn(total_timesteps=chunk_size, reset_num_timesteps=False)

    # ‚úÖ Overwrite same model file each time
    model.save(base_model_path)
    print(f"[üíæ Saved model to {base_model_path}.zip]")

# --- Final test pass ---
print("\n[üöÄ FINAL TESTING PHASE ‚Äî Single Pass Through Test Set]")
obs = test_env.reset()
total_rewards, episode_reward = [], 0.0

for step in range(len(test_df)):
    action, _ = model.predict(obs, deterministic=True)
    obs, reward, done, info = test_env.step(action)
    episode_reward += reward[0]

    if done[0]:
        total_rewards.append(episode_reward)
        print(f"Episode reward: {episode_reward:.2f}")
        episode_reward = 0
        obs = test_env.reset()

# --- Results summary ---
print("\n[üìä TESTING COMPLETE]")
if total_rewards:
    print(f"Average Reward: {np.mean(total_rewards):.2f}")
    print(f"Max Reward: {np.max(total_rewards):.2f}")
    print(f"Min Reward: {np.min(total_rewards):.2f}")
else:
    print("No episodes completed during test run.")
```
